```python
#  merging tables: 
# Start with edstays data - select specific columns
merged_data = data_edstays[['subject_id', 'stay_id', 'intime', 'outtime', 'race']]

# Add triage data for chief complaints
merged_data = pd.merge(
    merged_data,
    data_triage[['stay_id', 'chiefcomplaint']],
    on='stay_id',
    how='inner'
)

# Add patient demographics (note: subject_id will be used for merge but not duplicated)
merged_data = pd.merge(
    merged_data,
    patients[['subject_id', 'gender', 'anchor_age', 'anchor_year', 'anchor_year_group', 'dod']],
    on='subject_id',
    how='left'
)

# Select final features (subject_id appears only once, gender from patients will overwrite gender from edstays)
final_data = merged_data[['subject_id', 'stay_id', 'intime', 'outtime', 'gender', 'race', 'chiefcomplaint', 'anchor_age', 'anchor_year', 'anchor_year_group', 'dod']]
```

```python
# ============================================================================
# DATA CLEANING: HANDLING NEGATIVE DEATH DAYS
# ============================================================================
# 
# PROBLEM IDENTIFIED:
# During mortality analysis, we discovered 567 cases where patients appeared to 
# die BEFORE their ED admission (negative death days). This is medically impossible
# and indicates data quality issues in the MIMIC dataset.
#
# DETAILED ANALYSIS OF THE 567 NEGATIVE CASES:
# 
# 1. TIME PRECISION ISSUES (546 cases = 96.3%):
#    - These cases show exactly -1 day difference
#    - Pattern: ED admission has specific time (e.g., "2128-07-31 09:19:00")
#              Death date recorded as midnight (e.g., "2128-07-31 00:00:00")
#    - Explanation: Same-day deaths where death time wasn't recorded precisely,
#      defaulting to 00:00:00, making it appear to occur before admission
#    - Solution: Convert these to 0 days (same-day death)
#
# 2. MINOR DATE RECORDING ERROR (1 case = 0.2%):
#    - One case with -2 days difference
#    - Patient had 3-minute ED visit, death date appears to be off by 1 day
#    - Likely data entry error
#    - Solution: Remove due to uncertain data quality
#
# 3. MAJOR DATA INTEGRITY ERRORS (20 cases = 3.5%):
#    - Deaths recorded 24 to 3,091 days (up to 8.5 years) before ED admission
#    - Examples: ED visit in 2179, death recorded in 2171 
#    - Clearly impossible scenarios indicating database corruption, patient ID
#      mix-ups, or data processing errors during MIMIC dataset creation
#    - Solution: Remove these cases as they represent clear data corruption
#
# IMPACT OF CLEANING:
# - Fixes 546 cases (preserves data while correcting time precision issues)
# - Removes 21 cases (only 0.005% of total data) with clear data errors
# - Ensures reliable mortality calculations for downstream analysis
# - Maintains data integrity without significant data loss
#
# This approach balances data preservation with data quality, addressing the
# most likely explanation (time recording precision) while removing clearly
# erroneous cases that would skew mortality analysis results.
# ============================================================================

print("=== FINAL DATA CLEANING DECISION ===")

# Count original negative cases for tracking
original_negative_cases = (final_data_clean['dead_in_days'] < 0).sum()
print(f"Original negative cases: {original_negative_cases}")

# Fix -1 day cases (time precision issues) - convert to same-day deaths
time_precision_issues = final_data_clean['dead_in_days'] == -1
final_data_clean.loc[time_precision_issues, 'dead_in_days'] = 0
print(f"Fixed {time_precision_issues.sum()} cases from -1 to 0 days (time precision)")
```

```python
# satandarizing the race into common usula categories

def map_race_standard(race):
    if pd.isna(race):
        return 'Other/Unknown/Not Reported'
    
    standard_race_mapping = {
        # White variations
        'WHITE': 'White',
        'WHITE - OTHER EUROPEAN': 'White',
        'WHITE - RUSSIAN': 'White',
        'WHITE - EASTERN EUROPEAN': 'White',
        'WHITE - BRAZILIAN': 'White',
        'PORTUGUESE': 'White',
        
        # Black variations
        'BLACK/AFRICAN AMERICAN': 'Black',
        'BLACK/AFRICAN': 'Black',
        'BLACK/CAPE VERDEAN': 'Black',
        'BLACK/CARIBBEAN ISLAND': 'Black',
        
        # Hispanic/Latino variations
        'HISPANIC/LATINO - PUERTO RICAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - DOMINICAN': 'Hispanic/Latino',
        'HISPANIC OR LATINO': 'Hispanic/Latino',
        'HISPANIC/LATINO - GUATEMALAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - HONDURAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - SALVADORAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - CUBAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - COLUMBIAN': 'Hispanic/Latino',
        'SOUTH AMERICAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - MEXICAN': 'Hispanic/Latino',
        'HISPANIC/LATINO - CENTRAL AMERICAN': 'Hispanic/Latino',
        
        # Asian variations
        'ASIAN - CHINESE': 'Asian',
        'ASIAN': 'Asian',
        'ASIAN - SOUTH EAST ASIAN': 'Asian',
        'ASIAN - ASIAN INDIAN': 'Asian',
        'ASIAN - KOREAN': 'Asian',
        
        # Native American/Alaska Native
        'AMERICAN INDIAN/ALASKA NATIVE': 'American Indian/Alaska Native',
        
        # Native Hawaiian/Pacific Islander
        'NATIVE HAWAIIAN OR OTHER PACIFIC ISLANDER': 'Native Hawaiian/Pacific Islander',
        
        # Multiple race
        'MULTIPLE RACE/ETHNICITY': 'Multiple Races',
        
        # Other/Unknown/Not Reported (all grouped)
        'OTHER': 'Other/Unknown/Not Reported',
        'UNKNOWN': 'Other/Unknown/Not Reported',
        'PATIENT DECLINED TO ANSWER': 'Other/Unknown/Not Reported',
        'UNABLE TO OBTAIN': 'Other/Unknown/Not Reported',
    }
    
    return standard_race_mapping.get(race, 'Other/Unknown/Not Reported')
```
